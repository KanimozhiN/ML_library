### concatenation and append

```python
table1 ={'name':['dylan','darren','aron','steven'],
        'test1':[100,99,98,100],
        'test2':[100,100,98,99],
        'dept':['IT','cse','IT','mech']
        }
df1=pd.DataFrame(table1)
df1
```

```python
table2 ={'name':['goon wo','lee','kim','jeon'],
        'test1':[100,99,98,100],
        'test2':[100,100,98,99],
        'dept':['IT','mech','IT','eee']
        }
df2=pd.DataFrame(table2)
df2
```

```python
df=pd.concat([df1,df2])
df
```

```python
df1.append(df2)
```

### Joins

```python
table1 ={'name':['dylan','darren','aron','steven'],
        'cust_id':[100,101,102,103]
        }
         
    
df_cust1=pd.DataFrame(table1)
df_cust1
```

```python
table2 ={'id':[1,2,3,4],
         'order':[525,644,545,545],
        'cust_id':[100,101,102,103]
        }
df_cust2=pd.DataFrame(table2)
df_cust2
         
    
```

```python
df_cust1.set_index('cust_id').join(df_cust2.set_index('cust_id'), on='cust_id', how='inner')
#inner joins
```

```python
df_cust1.join(df_cust2,lsuffix='_customer',rsuffix='_order')
```

### Merge

```python
pd.merge(df_cust1,df_cust2,on='cust_id')
```

```python
pd.merge(df_cust1,df_cust2,on='cust_id',how ='outer')
```

```python
pd.merge(df_cust1,df_cust2,on='cust_id',how ='right')
```

```python
pd.merge(df_cust1,df_cust2,on='cust_id',how ='left')
```

```python
pd.merge(df_cust1,df_cust2,left_index=True,right_index=True)
```

### Reshaping

```python
details ={'name':['dylan','darren','aron','steven'],
          'city':['shengai','tianjin','beijing','chongqing'],
        'age':[24,25,27,26],
         'gender':["M","M","M","M"]
        }
df_det=pd.DataFrame(details)
df_det
```

```python
df_melt=df_det.melt(id_vars=['gender'])
df_melt
```

```python
df_melt=df_det.melt(id_vars=['gender'],value_vars='age')
df_melt
```

### pivot table

```python
pd.pivot_table(df_det, index=['name'], values=['age'])
```

```python
pd.pivot_table(df_det, index=['name'], values=['age'],aggfunc='sum')
```

### Cross Tables

```python
pd.crosstab(df_det.name,df_det.age,rownames=['name'],colnames=['age'])
```

```python
pd.crosstab(df_det.name,df_det.age,rownames=['name'],colnames=['age'], values=df_det['age'],aggfunc='mean')
```
